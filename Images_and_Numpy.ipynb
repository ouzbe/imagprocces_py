```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open("Honey_Badger.jpg")
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
```

```python
pic_red.shape
```

```python
#R G B

#Red channel
plt.imshow(pic_red[:,:,0],cmap='gray')
```

```python
#Green channel
plt.imshow(pic_red[:,:,1],cmap='gray')
```

```python
#Blue channel
plt.imshow(pic_red[:,:,2],cmap='gray')
```

```python
#Green channel
pic_red[:,:,1] = 0
#Remove the green channel
```

```python
pic_red[:,:,2] = 0
#Remove the blue channel
```

```python
plt.imshow(pic_red)
```
